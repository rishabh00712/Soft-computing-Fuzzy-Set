```python
import matplotlib.pyplot as plt
import numpy as np
C=np.linspace(-20,40,100)
Cy=[]
D=np.linspace(15,75,100)
Dy=[]
for i in C:
  if i>=-10 and i<0:
    Cy.append((i-(-10))/(0-(-10)))
  elif i>=0 and i<5:
    Cy.append(1.0)
  elif i<-10 or i>=5:
    Cy.append(0.0)
for i in D:
  if i>=25 and i<35:
    Dy.append((i-25)/(35-25))
  elif i>=35 and i<40:
    Dy.append(1.0)
  elif i<25 or i>=40:
    Dy.append(0.0)
fig, axs = plt.subplots(1, 2, figsize=(10, 5))


ax=axs[0]
ax.plot(C,Cy)
axs[0].set_title('Membership Function of C')
ax.set_xlabel('Celsius')

ax=axs[1]
ax.plot(D,Dy)
axs[ 1].set_title('Membership Function of D')
ax.set_xlabel('Celsius')
plt.tight_layout()

```

```python
import matplotlib.pyplot as plt

# Assuming C, D, Cy, and Dy are defined arrays/lists
R_CxD = []

for i in range(len(C)):
    r = []
    for j in range(len(D)):
        r.append(min(Cy[i], Dy[j]))
    R_CxD.append(r)

C_maxmin = []
D_maxmin = []
C_maxpro = []
D_maxpro = []

# Max-min composition for C and D
for j in range(len(D)):
    maxmin = []
    for i in range(len(C)):
        maxmin.append(min(R_CxD[i][j], Cy[i]))
    C_maxmin.append(max(maxmin))

for j in range(len(C)):
    maxmin = []
    for i in range(len(D)):
        maxmin.append(min(R_CxD[i][j], Dy[i]))
    D_maxmin.append(max(maxmin))

# Plotting max-min composition
fig, axs = plt.subplots(1, 2, figsize=(15, 7))

axs[0].plot(C, C_maxmin)
axs[0].set_title('Max-min composition of C and CxD')
axs[0].set_xlabel('Celsius')
axs[0].set_ylabel('Max-min Value')
axs[0].grid(True)

axs[1].plot(D, D_maxmin)
axs[1].set_title('Max-min composition of D and CxD')
axs[1].set_xlabel('Celsius')
axs[1].set_ylabel('Max-min Value')
axs[1].grid(True)

plt.tight_layout()
plt.show()

# Max-product composition for C and D
for j in range(len(D)):
    maxpro = []
    for i in range(len(C)):
        maxpro.append(R_CxD[i][j] * Cy[i])
    C_maxpro.append(max(maxpro))

for j in range(len(C)):
    maxpro = []
    for i in range(len(D)) :
        maxpro.append(R_CxD[i][j] * Dy[i])
    D_maxpro.append(max(maxpro))

# Plotting max-product composition
fig, axs = plt.subplots(1, 2, figsize=(15, 7))

axs[0].plot(C, C_maxpro)
axs[0].set_title('Max-product composition of C and CxD')
axs[0].set_xlabel('Celsius')
axs[0].set_ylabel('Max-product Value')
axs[0].grid(True)

axs[1].plot(D, D_maxpro)
axs[1].set_title('Max-product composition of D and CxD')
axs[1].set_xlabel('Celsius')
axs[1].set_ylabel('Max-product Value')
axs[1].grid(True)

plt.tight_layout()
plt.show()
```
