```python
import numpy as np
import matplotlib.pyplot as plt

def trapezoidal(x, a, b, c, d):
    if x <= a or x >= d:
        return 0
    elif a < x <= b:
        return (x - a) / (b - a)
    elif b < x <= c:
        return 1
    elif c < x < d:
        return (d - x) / (d - c)

# Define the range of x values
x_values = np.linspace(0, 20, 100)
y_values = [trapezoidal(x, 0, 9, 11, 20) for x in x_values]


plt.plot(x_values, y_values, label='Trapezoidal')
plt.title("Trapezoidal Membership Function")
plt.xlabel("x")
plt.ylabel("Membership Degree")
plt.grid(True)
plt.legend()
plt.show()
```

```python


def triangular(x, a, b, c):
    if x <= a or x >= c:
        return 0
    elif a < x <= b:
        return (x - a) / (b - a)
    elif b < x < c:
        return (c - x) / (c - b)

# Define the range of x values
x_values = np.linspace(0, 8, 100)
y_values = [triangular(x, 2.5, 4, 5.5) for x in x_values]

# Plotting the function
plt.plot(x_values, y_values, label='Triangular')
plt.title("Triangular Membership Function")
plt.xlabel("x")
plt.ylabel("Membership Degree")
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def gaussian(x, c, sigma):
    return np.exp(-((x - c) ** 2) / (2 * sigma ** 2))

# Define the range of x values
x_values = np.linspace(0, 8, 100)
y_values = gaussian(x_values, 4, 1)

# Plotting the function
plt.plot(x_values, y_values, label='Gaussian')
plt.title("Gaussian Membership Function")
plt.xlabel("x")
plt.ylabel("Membership Degree")
plt.grid(True)
plt.legend()
plt.show()
```
